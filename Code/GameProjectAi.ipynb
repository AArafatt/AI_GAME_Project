```python
import random
import itertools

def arithmetic_challenge():
    print("Human and the AI will compete to reach a target number using arithmetic expressions.")
    print("You can choose two numbers from the available list and use the four basic operators: +, -, *, /")
    print("The player who comes closest to the target number wins.")

    # Here Generate a list of 16 random numbers
    number_list = random.sample(range(1, 17), 16
                                )

    # Number of rounds
    num_rounds = 3

    human_score = 0
    ai_score = 0

    for round in range(1, num_rounds + 1):
        print("\n--- Round", round, "---")
        print("Available Numbers:", number_list)

        # Generate a random target number
        target = random.randint(1, 100)
        print("Target number:", target)

        # Human player's turn
        player_expression = get_player_expression(number_list)
        player_result = evaluate_expression(player_expression)
        print("Your expression:", player_expression, "= ", player_result)

        # AI player's turn
        ai_expression = get_ai_expression(number_list, target)
        ai_result = evaluate_expression(ai_expression)
        print("AI's expression:", ai_expression, "= ", ai_result)

        player_diff = abs(target - player_result)
        ai_diff = abs(target - ai_result)

        if player_diff < ai_diff:
            print("You win this round!")
            human_score += 1
        elif ai_diff < player_diff:
            print("AI wins this round!")
            ai_score += 1
        else:
            print("It's a tie this round!")


        player_num1, player_num2 = map(int, player_expression.split()[::2])
        ai_num1, ai_num2 = map(int, ai_expression.split()[::2])


        number_list.remove(ai_num1)
        number_list.remove(ai_num2)

    print("\n--- Final Scores ---")
    print("Your score:", human_score)
    print("AI score:", ai_score)

    if human_score > ai_score:
        print("Congratulations! You win the game!")
    elif ai_score > human_score:
        print("AI wins the game!")
    else:
        print("It's a tie!")

def get_player_expression(number_list):
    print("Player's Turn:")
    num1, num2 = get_player_numbers(number_list)
    operator = input("Enter an arithmetic operator (+, -, *, /): ")
    expression = f"{num1} {operator} {num2}"
    return expression

def get_player_numbers(number_list):
    num1 = get_number_choice("Choose a number from the available list: ", number_list)
    number_list.remove(num1)
    num2 = get_number_choice("Choose another number from the available list: ", number_list)
    number_list.remove(num2)
    return num1, num2

def get_number_choice(prompt, number_list):
    print("Available Numbers:", number_list)
    num = int(input(prompt))
    while num not in number_list:
        num = int(input("Invalid choice. Please select a number from the available list: "))
    return num

def get_ai_expression(number_list, target):
    print("AI's Turn:")
    expressions = []
    for num1, num2 in itertools.combinations(number_list, 2):
        for operator in ['+', '-', '*', '/']:
            expression = f"{num1} {operator} {num2}"
            expressions.append(expression)

    optimal_expression = None
    optimal_diff = float('inf')

    for expression in expressions:
        result = evaluate_expression(expression)
        if result is not None:
            diff = abs(target - result)
            if diff < optimal_diff:
                optimal_diff = diff
                optimal_expression = expression

    return optimal_expression

def evaluate_expression(expression):
    try:
        result = eval(expression)
        return result
    except:
        print("Invalid expression!")
        return None

# Run the game
arithmetic_challenge()
```
